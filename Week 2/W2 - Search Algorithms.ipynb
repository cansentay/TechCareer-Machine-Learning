## Linear Search

```python
def linear_arama(liste, hedef):
    for i in range(len(liste)):
        if liste[i] == hedef:
            return i
    return -1

Liste = [4, 2, 8, 19, 18, 63, 49, 76]
Aranan = 18

Sonuç = linear_arama(Liste, Aranan)

if Sonuç != -1:
    print(f"{Aranan} liste içinde {Sonuç}. indexte bulundu")
else:
    print(f"{Aranan} liste içerisinde bulunamadı")
```

## Binary Search

```python
def ikili_arama(liste,hedef):
  Sol = 0
  Sağ =len(liste) - 1

  while Sol<=Sağ:
    Orta = (Sol+Sağ)//2
    Orta_eleman = liste[Orta]
    if Orta_eleman == hedef:
      return Orta
    elif Orta_eleman<hedef:
      Sol = Orta + 1
    else:
      Sağ = Orta - 1
  return -1
Sırali_liste=[1,2,3,4,5,6,7,8,9]
Aranan = 9
Sonuç = ikili_arama(Sırali_liste,Aranan)
if Sonuç != -1:
    print(f"{Aranan} liste içinde {Sonuç}. indexte bulundu")
else:
    print(f"{Aranan} liste içerisinde bulunamadı")
```
