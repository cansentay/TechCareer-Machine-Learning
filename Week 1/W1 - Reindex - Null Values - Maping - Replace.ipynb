```python
from numpy import nan as NA
```

```python
dtf=pd.DataFrame([[1,2,3],[4,NA,5],[NA,NA,NA]])
dtf
```

```python
dtf.dropna(how="all") # Sadece tüm satır NAN ise siler
```

```python
dtf[1]=NA
```

```python
dtf
```

```python
dtf.dropna(axis=1,how="all")
```

```python
dtf.dropna(thresh=1) #Satırda en az 1 tane eksik olmanayan değer varsa silmez
```

```python
dtf.dropna(thresh=3)
```

```python
dtf
```

```python
dtf.fillna(17)
```

```python
dtf.fillna({0:29,1:38,2:76}) # Sütunlara göre eksik değer ataması
```

```python
dtf
```

```python
dtf.fillna(method="ffill") # Üs satırdaki değerle doldurur
```

```python
dtf.fillna(method="ffill",limit=1) # Sütundaki sadece tek bir NAN Doldurur
```

```python
series=pd.Series([1,0,NA,9])
series
```

```python
series.fillna(series.mean()) #Tüm Seriyi tüm serninin ortalamasıyla doldurur
```

```python
dtf.fillna(dtf.mean())#Tüm Dataframe'i tüm serninin ortalamasıyla doldurur
```

### Maping

```python
vr2=pd.DataFrame({"isim":["Ali","Mehmet","Nur","Fatma"],
                  "Notlar":[90,12,85,100]})
```

```python
vr2
```

```python
siniflar={"Ali":"A","Mehmet":"B","Nur":"C","Fatma":"A"}
```

```python
adlar=vr2["isim"]
```

```python
vr2["sube"]=adlar.map(siniflar)
```

```python
vr2
```
